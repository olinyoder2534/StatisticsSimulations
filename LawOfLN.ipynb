```python
import matplotlib.pyplot as plt
import random
```

```python
#Coin toss
results = [random.choice(['Heads', 'Tails']) for i in range(5000)]

heads_count = 0
tails_count = 0
proportions = []

for j, result in enumerate(results, start=1):
    if result == 'Heads':
        heads_count += 1
    else:
        tails_count += 1
    proportion = heads_count / j
    proportions.append(proportion)
```

```python
#plot
plt.plot(range(1, len(results) + 1), proportions, marker='o')
#horizontal line at y = .5
plt.axhline(y=0.5, color='red', linestyle='--')
plt.xlabel('Num Tosses')
plt.ylabel('Proportion of Heads Tosses')
plt.title('Proportion of Heads to Total Tosses')
plt.show()
```

```python
print('Num Heads: {}'.format(heads_count))
print('Num Tails: {}'.format(tails_count))
```
